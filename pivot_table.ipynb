```python
df = result
df['DW'] = 0
def day_of_week(day):
    weekday = day.weekday()
    if weekday == 0 :
        return '2'

    elif weekday == 1 :
        return '3'

    elif(weekday == 2) :
        return '4'

    elif(weekday == 3) :
        return '5'

    elif(weekday == 4) :
        return '6'

    elif(weekday == 5) :
        return '7'

    elif (weekday == 6) :
        return 'CN'
df.DW = df.apply(lambda row: day_of_week(row['DATE']), axis = 1)
df.DATE = df.apply(lambda row: row['DATE'].date(), axis = 1)
df
```

```python
df = df[['CODE','NAME','POSITION','DEPARTMENT','DATE','DW','CHECKIN','CHECKOUT']]
df
```

```python
import datetime 
df['status'] = None
def get_status(start = '07:30', end = '17:30'):
    try:
        checkin = datetime.datetime.strptime(start, '%H:%M')
        checkout = datetime.datetime.strptime(end, '%H:%M')
        lunch_start = datetime.datetime.strptime('12:00', '%H:%M')
        lunch_end = datetime.datetime.strptime('13:30', '%H:%M')
        timedelta_obj = (checkout - checkin)
        diff_in_hours = timedelta_obj.total_seconds() / 3600
        if (diff_in_hours < 3.5):
            return 'T'
        elif (checkin > lunch_start):
            return 'X/2'
        elif (checkin < lunch_end) & (checkout < lunch_end):
            return 'X/2'
        else:
            return 'X'
    except:
        return 'T'
        
df.status = df.apply(lambda row: get_status(row['CHECKIN'],row['CHECKOUT']), axis = 1)
df
```

```python

df.loc[:,'islate'] = 0
def islate(start = '07:30'):
    if start == 0:
        return 0
    start = datetime.datetime.strptime(start, '%H:%M')
    fix_time = datetime.datetime.strptime('08:10', '%H:%M')
    timedelta_obj = (start - fix_time)
    diff_in_minutes = timedelta_obj.total_seconds() / 60
    if diff_in_minutes > 0 :
        return 1
    else: return 0
df.islate = df.apply(lambda row: islate(row['CHECKIN']), axis = 1)
df
```

```python
df['late_minutes'] = 0
def get_minutes_late(start = '07:30'):
    if start == 0:
        return 0
    start = datetime.datetime.strptime(start, '%H:%M')
    fix_time = datetime.datetime.strptime('08:10', '%H:%M')
    timedelta_obj = (start - fix_time)
    diff_in_minutes = timedelta_obj.total_seconds() / 60
    if diff_in_minutes > 0:
        return diff_in_minutes
    else: return 0
df.late_minutes = df.apply(lambda row: get_minutes_late(row['CHECKIN']), axis = 1)
df
```

```python
df['isearly'] = 0
def isearly(hour = '07:30'):
    if hour == '':
        return 0
    hour = datetime.datetime.strptime(hour, '%H:%M')
    end = datetime.datetime.strptime('17:30', '%H:%M')
    lunch_start = datetime.datetime.strptime('12:00', '%H:%M')
    lunch_end = datetime.datetime.strptime('13:30', '%H:%M')
    if hour > end:
        return 0
    elif hour < lunch_start:
        return 1
    elif lunch_start< hour < lunch_end:
        return 0
    elif hour < end:
        return 1
    else: 
        return 0
df.isearly = df.apply(lambda row: isearly(row['CHECKOUT']), axis = 1)
df
```

```python
df['early_minutes'] = 0
def get_early_minutes(hour):
    if hour == '':
        return 0
    
    hour = datetime.datetime.strptime(hour, '%H:%M')
    end = datetime.datetime.strptime('17:30', '%H:%M')
    lunch_start = datetime.datetime.strptime('12:00', '%H:%M')
    lunch_end = datetime.datetime.strptime('13:30', '%H:%M')
    if hour > end:
        return 0
    elif hour < lunch_start:
        return (lunch_start - hour).total_seconds() / 60
    elif lunch_start< hour < lunch_end:
        return 0
    elif hour < end:
        return (end - hour).total_seconds() / 60
    else: 
        return 0
df.early_minutes = df.apply(lambda row: get_early_minutes(row['CHECKOUT']), axis = 1)
df
```

```python
df['work_hours'] = 0
def get_work_hours(start,end):
    if end == '':
        return 0
    start = datetime.datetime.strptime(start, '%H:%M')
    end = datetime.datetime.strptime(end, '%H:%M')
    start_afternoon = datetime.datetime.strptime('12:00', '%H:%M')
    end_afternoon = datetime.datetime.strptime('13:30', '%H:%M')
    if (start_afternoon > start) and (end > end_afternoon):
        timedelta_obj = (end - start)
        diff_in_hours = timedelta_obj.total_seconds() / 3600
        return (diff_in_hours - 1.5)
    elif start_afternoon<start<end_afternoon:
        timedelta_obj = (end - end_afternoon)
        diff_in_hours = timedelta_obj.total_seconds() / 3600 
        return diff_in_hours
    elif start>end_afternoon:
        timedelta_obj = (end - start)
        diff_in_hours = timedelta_obj.total_seconds() / 3600 
        return diff_in_hours
    else:
        timedelta_obj = (end - start)
        diff_in_hours = timedelta_obj.total_seconds() / 3600
        return diff_in_hours
df.work_hours = df.apply(lambda row: get_work_hours(row['CHECKIN'],row['CHECKOUT']), axis = 1)
df
```

```python
def format_excel(file_name, start_date, end_date):
    import openpyxl
    from openpyxl.styles import Alignment,Font
    from string import ascii_uppercase
    work_book = openpyxl.load_workbook(filename = file_name) 
    
    work_sheet1 = work_book['general']
    work_sheet1.merge_cells('A1:C2')
    work_sheet1.cell(column=1, row=1, value=f" FROM {start_date} TO {end_date} ")
    work_sheet1.freeze_panes= 'C4'
    work_sheet1.cell(column=1, row=work_sheet1.max_row + 1, value="Note:")
    work_sheet1.cell(column=1, row=work_sheet1.max_row + 1 , value="X: Ngày đi làm (2 ca).")
    work_sheet1.cell(column=1, row=work_sheet1.max_row + 1 , value="X/2 : Đi làm nửa ngày.")
    work_sheet1.cell(column=1, row=work_sheet1.max_row + 1 , value="V: Ngày nghỉ")
    work_sheet1.cell(column=1, row=work_sheet1.max_row + 1 , value="T: Quên checkin/checkout")
    for row in range(1,work_sheet1.max_row+1):
        for col in range(1,work_sheet1.max_column+1):
            cell=work_sheet1.cell(row, col)
            cell.alignment = Alignment(horizontal='center', vertical='center',wrap_text=False, shrink_to_fit = False)
            cell.font = Font(name='Times New Roman',size=8,color='FF000000')


    
    work_sheet2 = work_book['details']
    work_sheet2.freeze_panes= 'C2'
    for row in range(1,work_sheet2.max_row+1):
        for col in range(1,work_sheet2.max_column+1):
            cell=work_sheet2.cell(row, col)
            cell.alignment = Alignment(horizontal='center', vertical='center',wrap_text=False, shrink_to_fit = False)    
            cell.font = Font(name='Times New Roman',size=8,color='FF000000') 
    work_book.save(file_name)
    return 0 
```

```python
def autofit_width(file_name):
    import xlwings as xw
    with xw.App(visible=False) as app:
        wb = xw.Book(file_name)
        for sheet in range(len(wb.sheets)):
            ws = wb.sheets[sheet]
            ws.autofit()
        wb.save()
        wb.close()
    return 0
```
